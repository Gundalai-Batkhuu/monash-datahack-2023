```python
# List of columns to exclude
exclude_columns = ['Tonnes']

# Create a dictionary to store unique values for each column
unique_values_dict = {}

# Loop through columns and get unique values
for column in df.columns:
    if column not in exclude_columns:
        unique_values_dict[column] = df[column].unique()

# Print unique values for each column
for column, unique_values in unique_values_dict.items():
    print(f"Unique values for {column}:")
    print(unique_values)
    print()
    
    
```

Attributes Cat. order, Sub-stream and Type order have leading characters for indicing the categories. Extracting these indices

```python
# Columns to clean
columns_to_clean = ['Cat. order', 'Sub-stream', 'Type order']

for column in columns_to_clean:
    
    # Split the values in 'column' by a space and keep only the first part
    df[column] = df[column].str.split(' ').str.get(0)
    
    # Convert the column's data type to integer
    df[column] = df[column].astype(float).astype(pd.Int32Dtype())  

df.head()
```

Attribute Year contains a time range values. These should be converted to datetime values instead. 

```python
# Function to split 'Year' and convert it to datetime
def split_and_convert(year_range):
    start_year, end_year = map(int, year_range.split('-'))
    start_date = datetime(start_year, 1, 1)
    end_date = datetime(end_year, 12, 31)
    return start_date, end_date

# Apply the function to create 'Start date' and 'End date' columns
df[['Start date', 'End date']] = df['Year'].apply(split_and_convert).apply(pd.Series)

df.head()
```

```python
df.info()
```

```python
# Removing commas and converting the Tonnes column's data type to integer
df['Tonnes'] = df['Tonnes'].str.replace(',', '').astype(int)
```

```python
# Removing columns with redundant values
df = df.drop(columns=['Year'])
```
